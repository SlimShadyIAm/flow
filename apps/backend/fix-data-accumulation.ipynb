```python
# these are the batches we need to correct
# 18's data is present in 999, and 999's data is present in 7, etc.
batches = [
  [98, 97, 99],
  # [18, 999, 7, 10, 16, 21],
  # [17, 19, 13, 12],
  # [9, 8]
]
```

```python
import json
import os
for batch in batches:
  seen_timestamps = set()

  for participant_id in batch:
    # read all the files under `eye_tracker_data`
    files = os.listdir("eye_tracker_data")
    matching_files = [f"eye_tracker_data/{_file}" for _file in files if f"[{participant_id}]-" in _file]
    matching_files.sort()

    for _file in matching_files:
      with open(_file, "r") as f:
        data = json.load(f)
        print("found file " + _file)

      gaze_data = data['data']
      data_to_write = []
      for line in gaze_data:
        if line['device_time_stamp'] not in seen_timestamps:
          seen_timestamps.add(line['device_time_stamp'])
          data_to_write.append(line)

      print(f"File {_file}: Removed {len(gaze_data) - len(data_to_write)} duplicates out of {len(gaze_data)}")
      data['data'] = data_to_write
    
      # write the data back to the file
      with open(f"eye_tracker_data_fixed/{_file.replace('eye_tracker_data/', '')}", "w") as f:
        json.dump(data, f)
```
